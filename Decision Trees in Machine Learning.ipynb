# Decision Trees in Machine Learning ?

- Decision Trees are a widely-used machine learning algorithm suitable for both classification and regression tasks. As a supervised learning method, they construct a model in the form of a tree structure. Each internal node represents a decision based on a feature, each branch represents the outcome of that decision, and each leaf node represents a class label or a predicted value.  

- The decision tree algorithm learns from training data by recursively partitioning it based on features, aiming to create homogeneous subsets at each internal node. The decision at each node is made based on a specific feature and threshold value, with the goal of maximizing information gain or minimizing impurity at each step.

- In classification tasks, decision trees predict the class label of a sample by traversing the tree from the root to a leaf node. Each internal node applies a decision rule, and each leaf node corresponds to a class label. In regression tasks, decision trees predict a continuous value by assigning the average value of the training samples in each leaf node.

- Decision trees have several advantages, including their interpretability, as the tree structure can be easily visualized and understood. They handle both categorical and numerical features and can capture complex relationships between variables. However, decision trees are prone to overfitting, especially when the tree depth is not properly controlled. This can be mitigated by using techniques like pruning or ensemble methods such as Random Forest or Gradient Boosting, which combine multiple decision trees to improve performance.

- Overall, decision trees are versatile and widely-used machine learning models that provide a clear and intuitive representation of decision-making processes in various domains.

**Advantages**:
- Easy to understand and interpret.
- Requires little data preprocessing.
- Can handle both numerical and categorical data.

**Disadvantages**:
- Prone to overfitting.
- Can be biased towards features with more levels.

# 4. Data Splitting

### Step 01 : Selecting the Target Variable

````python
y = dataset.loc[:,"Survived"]
````

**Purpose:** This command extracts the target variable (or dependent variable) from the dataset.

**Explanation:**
- `dataset`: This is the DataFrame that contains your entire dataset.
- `.loc`: This is a label-based indexer provided by pandas for selecting rows and columns.
- `[:, "Survived"]`: This specifies that we want to select all rows (`:`) and the column named "Survived".

### Step 02 : Selecting the Feature Variables

````python
X = dataset.drop("Survived", axis=1)
````

**Purpose:** This command creates a DataFrame that contains all the feature variables (or independent variables) by removing the target variable column.

**Explanation:**

- `test_size=0.3`: This parameter specifies that 30% of the data should be used for testing, and the remaining 70% will be used for training.


# 5. Model Training

### Step 01 : Initializing the Decision Tree Classifier

````python
decisiontree = DecisionTreeClassifier()
````

**Purpose:** This command initializes a new instance of the `DecisionTreeClassifier` from the scikit-learn library.

**Explanation:**
- `DecisionTreeClassifier()`: This creates an empty decision tree classifier object with default parameters.
- This classifier will be used to fit the model to the training data and make predictions.


### Step 02 : Training the Model

````python
decisiontree.fit(X_train, y_train)
````

**Purpose:** This command trains the decision tree classifier on the training data.

**Explanation:**
- `fit()` is a method that takes two arguments: `X_train` and `y_train`.
- `X_train`: This is the feature matrix of the training data. Each row represents a training sample, and each column represents a feature.
- `y_train`: This is the target vector of the training data. Each element corresponds to the target value (or class label) for the respective training sample in `X_train`.
- The `fit()` method builds the decision tree model by finding patterns in the training data that map the features to the target values.

### Step 03 : Making Predictions

````python
decisiontree_pred = decisiontree.predict(X_test)
````

**Purpose:** This command uses the trained decision tree model to make predictions on the test data.

**Explanation:**
- `predict()` is a method that takes one argument: `X_test`.
- `X_test`: This is the feature matrix of the test data. Similar to `X_train`, each row represents a test sample, and each column represents a feature.
- The `predict()` method uses the decision tree model to predict the target values for each sample in `X_test`.
- The predictions are stored in the variable `decisiontree_pred`, which will be a vector of predicted class labels corresponding to the samples in `X_test`.

# 8. Confusion Matrix

A confusion matrix is an essential tool for evaluating the performance of a classification algorithm. It provides a detailed summary of how well the model performs in terms of predicting both correct and incorrect classifications across different categories. This summary is structured in a tabular format.

In evaluating a classification model, four key outcomes are considered:

#### True Positives (TP):
- These occur when the model correctly predicts that an observation belongs to a certain class, and it actually does.
  
#### True Negatives (TN):
- These occur when the model correctly predicts that an observation does not belong to a certain class, and it indeed does not.
  
#### False Positives (FP):
- These occur when the model predicts that an observation belongs to a certain class, but it actually does not (Type I error).
  
#### False Negatives (FN):
- These occur when the model predicts that an observation does not belong to a certain class, but it actually does (Type II error).

These outcomes are succinctly summarized within the confusion matrix, which provides a comprehensive view of the model's performance and the types of errors encountered.
